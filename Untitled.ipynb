```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('cleaned_data.xlsx')
```

```python
df.head()
```

```python
df.columns
```

```python
data = df[['Best_Sleep_Imp', 'Constant_BT_Imp', 'Enough_Sleep_Imp',
       'Ideal_Env_Imp', 'Best_Sleep_Perf', 'Constant_BT_Perf',
       'Enough_Sleep_Perf', 'Ideal_Env_Perf',]]
data.head()
data1 = data.copy()
```

```python
data.head()
```

```python
data.to_excel('Imp_vs_Perf.xlsx')
```

```python
data.replace({"Strongly disagree":1,"Disagree":1,"Neutral":2,"Agree":3,"Strongly agree":3},inplace = True)
```

```python
plt.figure(figsize=(16,8))
plt.plot(data1['Best_Sleep_Imp'], label = "Importance of best sleep")
plt.plot(data1['Best_Sleep_Perf'], label = "Performace of best sleep")
```

```python
fig = px.line(data1, x = data1['Best_Sleep_Imp'])
fig += px.line(data1, y = 'Best_Sleep_Perf')
fig.show()
```

```python
# Best Sleep Imp VS Perf
plt.figure(figsize=(24,8))
plt.plot(data1['Best_Sleep_Imp'], '-r', label = 'Expected')
plt.plot(data1['Best_Sleep_Perf'],'o-g', label = 'Actual')
plt.title("Best Sleep Expectation VS Best Sleep Performance", loc = "center")
plt.xlabel("No of Individuals")
plt.ylabel("Scale of Agreement")
plt.legend(loc="upper right")
plt.show()
```
